üß† Migraine Symptom Dataset for Classification:


*üìò* Description:

A migraine is a neurological disorder characterized by recurrent episodes of moderate to severe headache, often accompanied by symptoms such as nausea, vomiting, and sensitivity to light and sound. The headache typically presents as a throbbing or pulsating pain, usually on one side of the head, and can last from a few hours to several days and can significantly impact daily life.

Dataset: [https://www.kaggle.com/datasets/gzdekzlkaya/migraine-symptom-dataset-for-classification](https://www.kaggle.com/datasets/gzdekzlkaya/migraine-symptom-dataset-for-classification)



üìä Dataset Features (Columns)

```python
symptom_cols = [
    'Nausea', 'Vomit', 'Phonophobia', 'Photophobia', 'Visual', 'Sensory',
    'Dysphasia', 'Dysarthria', 'Vertigo', 'Tinnitus', 'Hypoacusis', 'Diplopia',
    'Defect', 'Ataxia', 'Conscience', 'Paresthesia', 'DPF'
]
```

```python
# Sum the values for each symptom column across all rows in the DataFrame
symptom_counts = df2[symptom_cols].sum()

# Sort the summed values in descending order to see the most frequent symptoms first
symptom_counts = symptom_counts.sort_values(ascending=False)

# Print the sorted counts of each symptom
print(symptom_counts)
```

```python
# Create a new figure with a size of 10 inches wide by 5 inches tall
plt.figure(figsize=(10, 5))

# Create a bar plot using Seaborn
# x: symptom names (categories) from symptom_counts index
# y: corresponding counts (number of patients) from symptom_counts values
# palette: color scheme 'coolwarm' for the bars
sns.barplot(x=symptom_counts.index, y=symptom_counts.values, palette='coolwarm')

# Set the title of the plot
plt.title('Common Migraine Symptoms')

# Label the y-axis to indicate it represents the number of patients
plt.ylabel('Number of Patients')

# Label the x-axis to indicate it represents symptoms
plt.xlabel('Symptom')

# Rotate x-axis labels by 45 degrees for better readability
plt.xticks(rotation=45)

# Adjust subplot parameters to give specified padding and prevent overlap
plt.tight_layout()

# Display the final plot
plt.show()
```

The above chart illustrates the most common symptoms reported by migraine patients.
Visual disturbances are the most prevalent, affecting over 600 individuals, followed by nausea, photophobia (light sensitivity), and phonophobia (sound sensitivity), each affecting around 400 patients. Less frequent symptoms include dizziness, vomiting, sensory disturbances, and vertigo. Rare symptoms such as dysarthria, ataxia, and diplopia are reported by very few patients.

This highlights that while migraines commonly impact vision and sensory perception, more severe neurological symptoms are relatively uncommon.

To analyze how migraine symptoms vary across different types of migraines.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Group by migraine type and compute the mean for binary symptoms
symptom_by_type = df.groupby('Type')[symptom_cols].mean()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(symptom_by_type, annot=True, cmap="coolwarm", cbar_kws={'label': 'Symptom Frequency'})
plt.title("Symptom Frequency by Migraine Type")
plt.xlabel("Symptom")
plt.ylabel("Migraine Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The above heatmap shows the frequency of different symptoms across various migraine types. Each cell represents how often a particular symptom (like nausea, visual disturbances, vertigo, etc.) occurs for each migraine type. Darker red shades indicate higher frequency, while blue shades indicate lower or zero occurrence.


From the plot:


Visual symptoms are highly prevalent in "Typical aura without migraine" and "Basilar-type aura".


Common symptoms like nausea, phonophobia, and photophobia are present across almost all migraine types.


Less common symptoms such as ataxia, diplopia, and dysarthria appear very rarely or not at all.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, chi2_contingency, kruskal
import numpy as np

# Select columns for analysis
cols = ['Age', 'Duration', 'Frequency', 'Location', 'Character', 'Intensity']

# Convert 'Location' and 'Character' to categorical if they aren't
df2['Location'] = df2['Location'].astype('category')
df2['Character'] = df2['Character'].astype('category')

# 1Ô∏è‚É£ Correlation Heatmap (Numeric only: Age, Duration, Frequency, Pain_Intensity)
numeric_cols = ['Age', 'Duration', 'Frequency', 'Intensity']
corr_matrix = df2[numeric_cols].corr(method='spearman')

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Spearman Correlation: Numeric Variables")
plt.show()

# 2Ô∏è‚É£ Kruskal-Wallis: Categorical (Location/Character) vs Numeric (Duration/Frequency/etc)
print("Kruskal-Wallis Test Results:\n")
for cat in ['Location', 'Character']:
    for num in ['Age', 'Duration', 'Frequency', 'Intensity']:
        groups = [group[num].dropna() for name, group in df.groupby(cat)]
        stat, p = kruskal(*groups)
        print(f"{cat} vs {num}: H={stat:.2f}, p={p:.4f}")

# 3Ô∏è‚É£ Chi-Square Test: Location vs Character (both categorical)
contingency_table = pd.crosstab(df2['Location'], df2['Character'])
chi2, p, dof, ex = chi2_contingency(contingency_table)
print(f"\nChi-Square Test between Location and Character:\nChi2 = {chi2:.2f}, p = {p:.4f}")
```

Statistical analysis revealed that pain intensity varies significantly depending on both the location and character of the migraine (p < 0.0001). Additionally, a strong association exists between pain location and character (Chi2 = 655.28, p < 0.0001), suggesting that these two features are often interlinked in how migraines present.
